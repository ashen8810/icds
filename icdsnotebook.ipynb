```python
def find_amounts(text):
    amounts = re.findall(r'\d+\.\d{2}\b', text)
    floats = [float(amount) for amount in amounts]
    unique = list(dict.fromkeys(floats))
    return unique
```

```python
t = 0
tt = []

import json
path = "./ICDS-Mini-Hackathon-2021/data/results-all.json"
ff = open(path,"r")
data = json.load(ff)

while(True):
    
    image = cv2.imread("./ICDS-Mini-Hackathon-2021/data/img/{0:03}.jpg".format(t))
    image= cv2.resize(image,None,fx=.5,fy=0.5)
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    ad= cv2.adaptiveThreshold(gray,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,cv2.THRESH_BINARY,85,11)

    if t == 50:
        break
    t=t+1
    print(t)
    text = pytesseract.image_to_string(ad)
    amounts = find_amounts(text)
    amounts.insert(0,data[t]["total"])
    
    tt.append(amounts)
    
```
